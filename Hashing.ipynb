```python
#importing the required libraries

import hashlib
#hashlib provides most of the hashing algorithms

import binascii
#binascii provides various tools to encode or decode and express in differernt 
#values

!pip install whirlpool
import whirlpool
#whirlpool provides a type of hashing.

#function program

def hash_01(x1):
  r = hashlib.md5(x1.encode())
  print(r.hexdigest())          #here the hexdigest() func will print the output in hexadecimal form

def hash_02(x2):
  n = b'x2'                     #Here we are converting the string value to byte value
  r1 = hashlib.md5(x2.encode())
  r2 = hashlib.sha1(x2.encode())
  r3 = whirlpool.new(n)
  print("md5 Hash : ", r1.hexdigest())
  print("sha1 Hash : ", r2.hexdigest())
  print("whirlpool Hash : ", r3.hexdigest())

def hash_03(x3):
  #pbkdf2_hmac(...)
  # pbkdf2_hmac(hash_name, password, salt, iterations, dklen=None) -> key
        
  #Password based key derivation function 2 (PKCS #5 v2.0) with HMAC as
  #pseudorandom function.
  hst1 = b'hst'
  rslt = hashlib.pbkdf2_hmac('sha256', hst1, b'saltthepassword', 100)
  print(binascii.b2a_hex(rslt))

def end_prog():
  input("Press Enter to continue....")

#Main Program
hst = input("Enter a string value for hashing : ")

while True:
  print("Operations for Hashing")
  print("1. To generate md5 Hash only")
  print("2. To generate Hash using 3 different algorithms [md5, sha1, whirlpool]")
  print("1. To generate a Hash and add Iterations+salting ")
  print("1. To Exit")

  ch = int(input("Enter your choice (1-3)"))
  if ch == 1:
    print("The Hsah of the given string is :")
    hash_01(hst)
    end_prog()

  elif ch == 2:
    print("The Hsah of the given string is :")
    hash_02(hst)
    end_prog()

  elif ch == 3:
    print("The Hsah of the given string is :")
    hash_03(hst)
    end_prog()

  elif ch == 4:
    print("Terminating Hashing program !!!!")
    print("thankyou for using the program")
    break

  else:
    print("Invalid input ?")
```
